### Ordinary least squares

- In statistics, ordinary least squares (OLS) is a type of linear least squares method for choosing the unknown parameters in a linear regression model (with fixed level-one effects of a linear function of a set of explanatory variables) by the principle of least squares: minimizing the sum of the squares of the differences between the observed dependent variable (values of the variable being observed) in the input dataset and the output of the (linear) function of the independent variable.


- Geometrically, this is seen as the sum of the squared distances, parallel to the axis of the dependent variable, between each data point in the set and the corresponding point on the regression surfacethe smaller the differences, the better the model fits the data. The resulting estimator can be expressed by a simple formula, especially in the case of a simple linear regression, in which there is a single regressor on the right side of the regression equation.


- The OLS estimator is consistent for the level-one fixed effects when the regressors are exogenous and forms perfect colinearity (rank condition), consistent for the variance estimate of the residuals when regressors have finite fourth moments andby the GaussMarkov theoremoptimal in the class of linear unbiased estimators when the errors are homoscedastic and serially uncorrelated.
- Under these conditions, the method of OLS provides minimum-variance mean-unbiased estimation when the errors have finite variances. 
- Under the additional assumption that the errors are normally distributed with zero mean, OLS is the maximum likelihood estimator that outperforms any non-linear unbiased estimator.

     to find such m and b value , where the loss is minimised and we can obtain minima, 
     we need to take partial derivative of both loss function with respect to slope and derivative both . 

```python
class Simple_Linear_regression_OLS:
    
    def __init__(self):
        self.m = None
        self.b = None
        
    def fit(self,x_train,y_train):
        
        numerator = np.sum((x_train - x_train.mean())*(y_train- y_train.mean()))
        denominator  = np.sum((x_train - x_train.mean())**2)
        
        self.m = numerator / denominator 
        
        self.b = y_train.mean() - self.m * x_train.mean()
        
        
        print(self.m)
        print(self.b)
    
    def predict(self,x_test):
        return self.m * x_test  + self.b
        
```

    MAE :  mean absolute error

    MAE = (1/n)  * sum ( abs( yi - yi-hat ) )
    
    advantage : 
    
        this loss number will be in the same unit as output unit. 
        like if taget variable is in Million Rs , then MAE also wll be in same unit. 
        
        also this is robust to outliers.
        drawback is we are using module(absolute)  function. which is not differenciable. which is mendatory for optimization. thats the reaason MSE is majorly used as loss function . 
        
     
    
    
    MSE = (1/n) * sum([(yi - yi_hat)^2])
    
    biggest advantage is it is differentiable 
    disadvantage : unit becomes squared. so for interpretation we have to take a square root
    emphasis the outlier values by square.
    not robust to outliers.
    
    
    
    
    RMSE = square_root(MSE)
    
    same properties as MSE 
    benefit is same unit as target 
    disadvantages: not that robust to outliers 
    mainly RMSE used in deep learning 
    
    
    
    
    
      
        

    R2_score : 
    
    
    R2 score , compares the how better is ur regression line compare to target_mean line.
    that is also called coefficient of determination.
    in literature it is also called Goodness of fit . 
    
    R2score =  SSR/SST
            = 1- SSE/ 
                 SST
    
            = 1 - sum of square of distnace of y-predicted to y-observed
                  dvided by sum of squared distance of y to mean y value
                
    SSR : squared error in the regression line .   SST-SSE = SSR
    
    
    
    
    R2_score = 1   -  (sum((yi-yi_hat)^2))/
                     (sum((yi-yi_bar)^2))
    
    
    
    interpretation of R2 score:
    
        lets say r2_score = 0.80
         that menas ur input variables are able to explain 80% of variance in target variable(output). , why that variance is coming out. 
         
        
    problem with R2 score : 
        if we add more input columns , the more variance will be explained by r2score. 
        
        to handle such behaviour we need adjust-r2score: 
        
        so , adjusted_R2_score = 1- (   (1- r2_score)(n-1)/
                                               (n-1-k))
                                                
                                               n = number if rows
                                               k = number of independent variables. 
        
        
        
        so in cases we have more columns , and some of the features are not very helpful for prediction , in such case we adjusted r2 score is more useful

### implementing mutliple linear regression from scratch  

    here , from above equation of finding all the betas.
    we need to do inverse operation of a matrix : which has O(n^3) time complexity:
    so , to optimize the algorithm  we use gradient descent .

```python
class MultipleLinearRegressor:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self,X_train,y_train):
        
        X_train = np.insert(X_train,0,1,axis = 1)
        
        # getting  beta0 and coefficients 
        
        all_beta = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        
        
        self.intercept_ = all_beta[0]
        self.coef_ = all_beta[1:]
        print("Intercept: ", self.intercept_)
        print("Coefficients: ", self.coef_)
        
    def predict(self,X_test):
        
        return  np.dot(X_test,self.coef_)+ self.intercept_
```

```python
MLR = MultipleLinearRegressor()
```

```python
MLR.fit(X_train,y_train)
```

```python
y_pred = MLR.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
